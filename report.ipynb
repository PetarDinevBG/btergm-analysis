```r
model_names <- c("btergm_full_","btergm_no_edges_","btergm_no_edges_nodematch_")
treatments <- c("Endow","Wealth + Endow","No info", "Wealth")
rownames(siena_results) <- c("nodecov.outdegree", "mutual", "ttriple","nodeicov.endowment","nodeocov.endowment", "nodeicov.wealth","nodeocov.wealth","int.  endowment alter x reciprocity","int.  wealth alter x reciprocity")

#Compare new Siena model with the three btergm models
for(t in treatments){
    siena_results <- read.csv(paste('Siena_output/params_', t, '.csv', sep=''))
    rownames(siena_results) <- c("nodecov.outdegree", "mutual", "ttriple","nodeicov.endowment","nodeocov.endowment", "nodeicov.wealth","nodeocov.wealth","int.  endowment alter x reciprocity","int.  wealth alter x reciprocity")
    comparison_matrix = matrix(0L, nrow = nrow(siena_results), ncol = 5)
    rownames(comparison_matrix) <- rownames(siena_results)
    colnames(comparison_matrix) <- c("Sign_Match", "Diff","Sign_Diff", "1_Insign", "2-Insign")
    for(n in model_names){
        file_name <- paste('btergm_output/',n, t, '.csv', sep='')
        btergm1_results <- read.csv(file_name,row.names = 1)
        for(row in rownames(siena_results)){
            if(row %in% rownames(btergm1_results)){
                if(sign(siena_results[row,4]) == sign(siena_results[row,5])){
                    if(sign(siena_results[row,2]) == sign(btergm1_results[row,1])){
                    comparison_matrix[row,1] <- comparison_matrix[row,1]+1
                    abs_diff <- abs(abs(siena_results[row,2]) - abs(btergm1_results[row,1]))
                    comparison_matrix[row,2] <- comparison_matrix[row,2]+abs_diff  
                    }else{
                    comparison_matrix[row,3] <- comparison_matrix[row,3]+1
                    }
                }else{
                    #TODO
                    comparison_matrix[row,4] <- comparison_matrix[row,4]+1
                }
                
            }
        }
    }
    print(t)
    print(comparison_matrix)
}
```
